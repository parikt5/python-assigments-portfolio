```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x,y=sp.symbols('x y')

f = x**2 + sp.sin(y)
# Differentiate f with respect to x
df_dx = sp.diff(f, x)
# Differentiate f with respect to y
df_dy = sp.diff(f, y)

#Define function
f = sp.exp(x) * sp.sin(y) + y**3
# Differentiate f with respect to x
df_dx = sp.diff(f, x)
# Differentiate f with respect to y
df_dy = sp.diff(f, y)
print ("the derivative of f with respect to x is:")
df_dx
```

```python
print ("the derivative of f with respect to y is:")
df_dy
```

```python
# Define the function g
g = x**2*y + y**2*x

# Compute partial derivatives (gradient components)
dg_dx = sp.diff(g, x) # Partial derivative with respect to x
dg_dy = sp.diff(g, y)  # Partial derivative with respect to y

#display gradient as a matrix (x on top, y on bottom)
sp.Matrix([dg_dx,dg_dy])

```

```python
#sub in value (1, -1) into gradient
point = {x: 1, y: -1}
value_at_x = dg_dx.subs(point)
value_at_y = dg_dy.subs(point)

print (value_at_x, ",", value_at_y)
```

```python
#define the function
h = sp.log(x**2 + y**2)
#finding the second mixed and unmixed partials
second_derivative_x = sp.diff((sp.diff(h, x)), x)
second_derivative_y = sp.diff((sp.diff(h, y)), y)
second_derivative_xy = sp.diff((sp.diff(h, x)), y)
second_derivative_yx = sp.diff((sp.diff(h, y)), x)
#can show them as a matrix
sp.Matrix([[second_derivative_x, second_derivative_y],[second_derivative_xy, second_derivative_yx]])
```

```python
print('Computing the second mixed partial derivative of g first with respect to x and then y gives the same result as differentiating first with respect to y and then x. This holds because the function is symmetric in x and y, meaning swapping the two variables does not alter the function. Consequently, the order of differentiation does not affect the outcome.')
```

```python
#defining the function
x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

#converting sympy to numerical function
j_func = sp.lambdify((x, y), j, 'numpy')

#creating a grid of x and y values and evaluate j on this grid
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

#plotting the contour:
plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

```python

# Define variables
x, y = sp.symbols('x y')

# Define a new function
k = sp.sin(x) * sp.cos(y) + (x**2)/2 - (y**2)/2

# Convert sympy function to a numerical function
k_func = sp.lambdify((x, y), k, 'numpy')

# Create a grid of x and y values
x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

# Plot the contour
plt.contourf(X, Y, Z, levels=50, cmap='plasma')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = sin(x) cos(y) + \\frac{x^2}{2} - \\frac{y^2}{2}$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```
